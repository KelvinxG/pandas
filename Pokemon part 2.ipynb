<b style="font-size:20px">Filtering data</b>
<li>Find string in columns 'Name' if it's contains 'Mega' </li>
<li> df.loc function , in short df.loc function will allow you to access a group of dataframe series ROWS and COLUMNS  <a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.loc.html"> here</a> </li>

```python
df=pd.read_csv("pokemon_data.csv")

#new_df=df.loc[(df['Type 1']== 'Grass') and (df['Type 2']=='Poison') and (df['HP']>70)]

df.loc[df['Name'].str.contains('Mega')]
```

<b style="font-size:20px">Regex filtering</b>
<li>use re librabry</li>




<li> re is a built-in python function : full info url <a href="https://docs.python.org/3/library/re.html"> here</a> </li>
<li>pandas .str.contains full info <a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.str.contains.html">  here</a> </li>


```python
import re

df.loc[df['Type 1'].str.contains('Fire|Grass',flags=re.I,regex=True)]
```

<b style="font-size:20px">Conditional changes</b>
<li> Inserting new a column 'Total' to combied all sort of numbers</li>
<li> Filter if Total is more than 500 in columns 'Generation','Legendary' will return 'YES YES YES'</li>

```python

#making changes by adding a column name 'Total'
df['Total']=df['HP']+ df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']+ df['Generation'] + df['Legendary']

df.head(5)
```

```python
df.loc[df['Total']>500,['Generation','Legendary']]='YES YES YES'
df
```

<b style="font-size:20px">Aggregate Statistics(Groupby)</b>
<li> In brief Groupby function allow you to shrink your data into smaller therefore you can compute operations on selected group </li>
<li> Groupby function  <a href="https://pandas.pydata.org/docs/reference/api/pandas.Series.groupby.html" >here</a> </li>

```python
df['count']=1

#df.groupby(['YOUR','COLUMNS']).operations such as .count() , .mean(), .sum() etc
#check parameters and operations in Groupby function above 

df.groupby(['Type 1','Type 2']).count()['count']
```

```python
new_df=pd.DataFrame(columns=df.columns)



for df in pd.read_csv('pokemon_data.csv',chunksize=5):
    results=df.groupby(['Type 1']).count()
    
    new_df=pd.concat([new_df,results])
print(new_df)
```
